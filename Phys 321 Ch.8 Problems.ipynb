```python
%pylab notebook
r = linspace(0.001,10,100)
n = 1
k = 20
u1 = k * r**n
u2 = 1/(2*r**2)
u = u1+u2
plt.plot(r,u1, '-.', r,u2, "b--", r,u)
plt.legend(["U1", "U2", "U"]);
plt.axis((0,10,-20,20))
```

```python
%pylab inline
r = linspace(0.001,1,100)
n = -1
k = 20
u1 = -k * r**n
u2 = 1/(2*r**2)
u = u1+u2
plt.plot(r,u1, ':', r,u2, "b--", r,u)
plt.legend(["U1", "U2", "U"]);
plt.axis((0,1,-200,100))
```
