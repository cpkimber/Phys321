In this problem, we are referred to an example presented earlier in the chapter that had been solved analytically, and promted to numerically integrate it using differential equation solving software.

```python
%pylab inline
from scipy.integrate import odeint
```

The equation for a damped-driven oscillator is given as 

$$\ddot{x} + 2 \beta \dot{x} + \omega_0^2 x = f(t)$$

Where 

$$f(t) = f_0 \cos(\omega t)$$

So the acceleration is given by

$$\ddot{x} = - 2 \beta \dot{x} - \omega_0^2 x + f_0 \cos(\omega t)$$

I begin by defining the variables as outlined in Example 5.3 on pg. 185.

```python
w  = 2*np.pi                       # Radians
w0 = 5*w                           # Radians
B  = w0/20                         # Radians
f0 = 1000                          # Meters
```

The next cell defines and calls the function dX/dt along the interval $0 < x < 5$, and solves the initial value problem with $x_0 = 0$ and $v_0 = 0$

```python
def dX_dt(X,t):
    return[X[1], -2*B*X[1] - w0**2 * X[0] + f0*np.cos(w*t)]

X0 = [0, 0]                        # Boundary conditions x_0 = 0 and v_0 = 0
ts = np.linspace(0, 5, 10000)      # Number array to show periods
Xs = odeint(dX_dt, X0, ts)         # Solution
ys = Xs[:, 0]                      # Solution array
```

```python
plt.plot(ts, ys)
pylab.xlim(xmin=0)
plt.xlabel("Time")
plt.ylabel("x(t)")
plt.title("Resulting Motion");
```

```python
plt.plot(ts, f0*np.cos(w*ts))
plt.xlabel("Time")
plt.ylabel("f0 cos(wt)")
plt.title("Driving Force");
```
