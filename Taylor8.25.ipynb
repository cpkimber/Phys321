```python
%pylab notebook
from scipy.optimize import fsolve
from scipy.integrate import odeint
```

```python
func = lambda rmin : (-2/3) * 1/rmin**(3/2) + 1/(2*rmin**2) + .1
rmin_guess=0.7
rmin_solution=fsolve(func,rmin_guess)
```

```python
r = linspace(0.001, 5, 10000)
u1 = -2/3 * r ** (-3/2)
u2 = 1/2 * r ** (-2)
u = u1 + u2
plt.plot(r,u1, r, u2, r,u, rmin_solution, -0.1, 'g*', markersize = 20)
plt.axis((0, 5, -0.2, 0.1))
plt.axvline(x = 1, ymin = -0.2, ymax = 5, linewidth = 2, linestyle = ":")
plt.axhline(y = -0.1, xmin = 0, xmax = 5, linewidth = 2, linestyle = "--")
plt.legend(["U1","U2","U", "r min", "Equilibrium", "Total Energy"])
plt.title("Problem 8.25")
plt.xlabel("Radius (m)")
plt.ylabel("Potential Energy (J)")
```
